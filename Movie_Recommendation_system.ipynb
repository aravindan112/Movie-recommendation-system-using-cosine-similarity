```python
import numpy as np
import pandas  as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

```

DATA COLLECTION AND PRE-PROCESSING

```python
#loading the dataset
movies_data = pd.read_csv('/content/movies dataset.csv')
```

```python
#printing the first 5 rows of the dataset to see a sample view
movies_data.head()
```

```python
movies_data.shape
```

```python
#selecting the relevant features for recommendation
selected_features = ['genres','keywords','tagline', 'cast', 'director']
print(selected_features)
```

```python

movies_data.isnull().sum()
```

```python
#replacing null values with a null string if there is null values in the selected features
for features in selected_features:
  movies_data[features] = movies_data[features].fillna('')
```

```python
movies_data.isnull().sum()
```

```python
#combining all the selected 5 features
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combined_features)
```

```python
#converting text data into feature vectors
vectroizer = TfidfVectorizer()
```

```python
feature_vectors = vectroizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

COSINE SIMILARITY

```python
#getting the similarity scores using cosine similarity
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
#getting the movie name from the user
movie_name = input('Enter your favourite movie name: ' )
```

```python
#creating an list with all the names of the movies given in the dataset
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#finding a similar or close match to the movie name that is given by the user input
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
#finding the index of the movie with title
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#getting a list of similar movies based on its index value
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
#sorting the movies based on its similarity score based on the users input movie
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)

```

```python
#print the name of similar movies based on the index
print('Movies Suggested for you : \n')
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  tital_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.', tital_from_index)
    i+=1
```

MOVIE RECOMMENDATION SYSTEM

```python
#getting the movie name from the user
movie_name = input('Enter your favourite movie name: ' )
list_of_all_titles = movies_data['title'].tolist()
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
close_match = find_close_match[0]
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
similarity_score = list(enumerate(similarity[index_of_the_movie]))
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print('Movies Suggested for you : \n')
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  tital_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.', tital_from_index)
    i+=1
```
